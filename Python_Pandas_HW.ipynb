```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
total_players = purchase_data.index.max() + 1
players_by_gender = purchase_data['Gender'].value_counts()
male_players = players_by_gender['Male']
female_players = players_by_gender['Female']
other_players = (total_players - male_players - female_players)

gender_demo = pd.DataFrame({
    "Gender" : ["Male", "Female", "Other/Non Disclosed"],
    "Total Count" : [male_players, female_players, other_players],
    "Percentage of Players" : [male_players/total_players, female_players/total_players, other_players/total_players]
})

gender_demo
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Get quartiles for binning
purchase_data.describe()

#set bins and labels
bins = [6, 20, 22, 25, 50]
binlabels = ['Under 21', '21 to 22', '23 to 25', '26 and Above']

#Add column for age bin
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins, labels = binlabels)

#Group by Age Group and count the total 

#Alternative Method:  
#age_group_demos = pd.DataFrame(purchase_data['Age Group'].value_counts())

age_group_demos = purchase_data.groupby('Age Group').agg({'SN': 'count'})

#Get total players
total_player_count = purchase_data.shape[0]

age_group_demos['Percent of Total'] = age_group_demos['SN']/total_player_count
age_group_demos["Percent of Total"] = age_group_demos["Percent of Total"].map("{:.2%}".format)
age_group_demos
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.head()

new_age_grouped = purchase_data.groupby('Age Group').agg({
    'Purchase ID': {'Count of Purchases': 'count'},
    'Price' : {'Avg Price' : 'mean', 'Total Purchases' : 'sum'},
    'SN' : {'Count of SN' : 'nunique'}
})
new_age_grouped.columns = new_age_grouped.columns.droplevel(0)

purchase_data.columns
new_age_grouped

#Add new columns
new_age_grouped['Avg Purchaces per Player'] = new_age_grouped['Count of Purchases']/new_age_grouped['Count of SN']
new_age_grouped['Avg PurchasePrice per Player'] = new_age_grouped['Total Purchases']/new_age_grouped['Count of SN']

#Format columns
new_age_grouped['Avg Price'] = new_age_grouped['Avg Price'].map("${:.2f}".format)
new_age_grouped['Total Purchases'] = new_age_grouped['Total Purchases'].map("${:.2f}".format)
new_age_grouped['Avg Purchaces per Player'] = new_age_grouped['Avg Purchaces per Player'].map("${:.2f}".format)
new_age_grouped['Avg PurchasePrice per Player'] = new_age_grouped['Avg PurchasePrice per Player'].map("${:.2f}".format)

new_age_grouped



```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data.groupby('SN').agg({
    'Purchase ID' : {'Purchase Count' : 'count'},
    'Price' : {'Avg. Purchase Price' : 'mean', 'Total Purchase Price' : 'sum'}
})

top_spenders.columns = top_spenders.columns.droplevel(0)
top_spenders_sorted = top_spenders.sort_values(by = ['Purchase Count'], ascending = False)
top_spenders_sorted['Avg. Purchase Price'] = top_spenders_sorted['Avg. Purchase Price'].map("${:.2f}".format)
top_spenders_sorted['Total Purchase Price'] = top_spenders_sorted['Total Purchase Price'].map("${:.2f}".format)
top_spenders_sorted
```
